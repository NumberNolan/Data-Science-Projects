```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
data = pd.read_csv('Salary_dataset.csv')
data.info()
data.head()
```

```python
x = data[['YearsExperience']]
y = data['Salary']

plt.tight_layout()
plt.scatter(x,y)
plt.xticks(size=14)
plt.yticks(size=14)
plt.xlabel('Years Experience', fontsize=14)
plt.ylabel('Salary ($)',fontsize=14)
plt.savefig('scatter.png')
plt.show()
```

```python
# split into test train 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .5, random_state = 16)

# train the model
model = LinearRegression()  
model.fit(x_train, y_train)
```

```python
# Print out the line
print("Intercept: ", model.intercept_)
print("Coefficient: ", model.coef_)
```

```python
y_pred = model.predict(x_test)
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test, y_pred, 'Red', label = 'Linear Regression Line')
plt.tight_layout()
plt.xticks(size=14)
plt.yticks(size=14)
plt.xlabel('Years Experience', fontsize=14)
plt.ylabel('Salary',fontsize=14)
plt.savefig('LinReg.png')
plt.legend()
plt.show()
```

```python
print('R squared: {:.2f}'.format(model.score(x_test,y_test)*100))
```
